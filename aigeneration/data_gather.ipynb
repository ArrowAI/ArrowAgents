# Define Variables

```python
categories = []
apps_ids = []
```

# Getting Categories

This was done manually and the categories were found from website

```python
with open('zapiercategories.txt') as zapiercategories:
    categories = zapiercategories.read().splitlines()

categories
```

# Getting Apps for the Categories

```python
import os
for category in categories:
    limit = 30
    appsurl = "https://zapier.com/api/v4/app-directory/"+category+"/apps?limit="+str(limit)+"&offset=0"

    while True:
        appresponse = requests.get(appsurl)
        if appresponse.status_code != 200:
            print("Error:", appresponse.status_code)

        apps = appresponse.json()
        for app in apps["results"]:
            if app["current_implementation_id"] == None:
                continue
            if apps_ids.count(app["current_implementation_id"]) == 0:
                apps_ids.append(app["current_implementation_id"])
                os.makedirs("data/"+category, exist_ok=True)
                with open("data/"+app["current_implementation_id"]+".json", "w") as f:
                    json.dump(app, f, indent=4)
                print("Added:", app["current_implementation_id"])
            else:
                print("Already Added:", app["current_implementation_id"])

        if apps["next"] == None:
            break
        appsurl = apps["next"]
        # len(apps["results"])
```

# Get List

```python
import requests
import json

allapps = "https://zapier.com/explore-api"

payload = {
    "operationName":"CategoryAppsBFFQuery",
    "variables":
    {
        "categorySlug":"all",
        "limit":22,
        "offset":0,
        "orderBy":"LAUNCHED",
        "filterBy":""
    },
    "query":"""
    query CategoryAppsBFFQuery($categorySlug: String = \"all\", $limit: Int = 10, $offset: Int = 0, $orderBy: AppSortOrder, $filterBy: String) {
        appCategory: appCategoryWithSlug(slug: $categorySlug) {
            id
            title
            apps(
                orderBy: $orderBy
            limit: $limit
            offset: $offset
            additionalCategorySlug: $filterBy
            ) {
                results {
                    id
                legacyId
                name
                logo {
                    mainUrl
                __typename
                }
                description
                isPremium
                isUpcoming
                isBeta
                slug
                profileUrl
                categories {
                    id
                title
                slug
                __typename
                }
                __typename
            }
            count
            hasNextPage
            __typename
            }
            featuredBlogPost {
                id
            title
            postUrl
            imageUrl
            apps {
                    id
                legacyId
                name
                slug
                description
                logo {
                    mainUrl
                __typename
                }
                primaryColor
                __typename
            }
            categories {
                    slug
                __typename
            }
            __typename
            }
            __typename
        }
    }
    
    """
}
headers = {'Content-Type': 'application/json'}

response = requests.post(allapps, headers=headers, data=json.dumps(payload))

if response.status_code != 200:
    print("Request Failed:",response.status_code)

allappsdata = response.json()
allappsdata["data"]["appCategory"]["apps"]["results"][0]
```
